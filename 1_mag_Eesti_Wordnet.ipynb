##Eesti Wordnet


Eesti Wordnet on leksikaal-semantiline andmebaas, mis sisaldab informatsiooni mõistete ja nendevaheliste semantiliste suhete kohta. Wordneti põhiühik on sünonüümihulk (ka sünohulk) ning võimalik on leida nii sõnade ülemmõisteid ehk hüperonüüme kui ka alammõisteid ehk hüponüüme.

```python
from estnltk.wordnet import wn #impordime vajalikud moodulid
from estnltk import Text    
```

```python
wn.all_synsets() #laeme alla kõik sünohulgad
```

```python
viisakas = wn.synsets("viisakas") #millistes sünohulkades on sõna "viisakas"
viisakas #selgus, et sõna "viisakas" sisaldub kahes sünohulgas
```

```python
viisakas1 = wn.synset('viisakas.a.01') #leiame sünonüümid esimeses sünohulgas
viisakas1.lemmas()
[lemma.name for lemma in viisakas1.lemmas()]
```

```python
viisakas2 = wn.synset('viisakas.a.02') #leiame sünonüümid teises sünohulgas
viisakas2.lemmas()
[lemma.name for lemma in viisakas2.lemmas()]
```

```python
for k in viisakas: #otsime ülemmõisteid
    print(k.hypernyms()) #ei leitud ühtegi
```

```python
for k in viisakas: #otsime alalmmõisteid
    print(k.hyponyms()) #ei leitud ühtegi
```
